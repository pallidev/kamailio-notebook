## Kamailio란?

**Kamailio**는 오픈소스 SIP 서버입니다. 전화 통화, 영상 통화, 메시징 등 실시간 통신의
핵심 프로토콜인 SIP(Session Initiation Protocol)를 처리합니다.

```
📱 전화기 → [SIP] → 🖥️ Kamailio → [SIP] → 📱 전화기
```

Kamailio의 동작은 **.cfg 파일**로 프로그래밍합니다. 이 노트북에서 그 cfg 코드를
셀 단위로 실행하면서 배워보겠습니다.

## 1. 첫 번째 변수 할당

Kamailio cfg에서 변수는 `$var(이름)` 형태로 사용합니다.

````python
$var(greeting) = "Hello, Kamailio!";
````

## 2. 숫자 변수

문자열뿐 아니라 정수도 저장할 수 있습니다. 타입은 자동으로 결정됩니다.

````python
$var(port) = 5060;
$var(port) = 5061;
$var(max_calls) = 1000;
````

## 3. 변수 출력하기

`xlog()` 함수로 변수를 포함한 메시지를 출력할 수 있습니다.

````python
xlog("Port: $var(port), Max calls: $var(max_calls)");
````

## 4. 변수 연결 (Concatenation)

`+` 연산자로 변수를 연결할 수 있습니다.

````python
$var(domain) = "example.com";
$var(user) = "alice";
$var(sip_uri) = "sip:" + $var(user) + "@" + $var(domain);
xlog("URI: $var(sip_uri)");
````

## 5. 모든 변수 확인하기

`%%vars` 매직으로 현재 설정된 모든 변수를 볼 수 있습니다.

````python
%%vars
````

## 6. 상태 초기화

`%%reset` 매직으로 모든 변수를 초기화할 수 있습니다.

````python
%%reset
````
