## What are Transformations Functions?

Transformations functions modify values by appending `{transform_name}` after a variable:

```
$(variable{transform_name})
```

Example: `$(ru{uri.user})` → extracts just the user part from `$ru`

## 1. URI Transformations

Extracting parts of a SIP URI is the most common transformation.

````kamailio-cfg
%%sip INVITE
From: <sip:1001@10.60.91.199:5060>;tag=abc
To: <sip:1002@10.60.80.218:5060>
````

````kamailio-cfg
$var(uri) = $ru;
xlog("Full URI: $var(uri)");
xlog("User part: $(var(uri){uri.user})");
xlog("Host part: $(var(uri){uri.host})");
xlog("Port part: $(var(uri){uri.port})");
````

## 2. Extract Caller ID from From URI

In production, you extract the caller ID for logging and routing decisions.

````kamailio-cfg
$var(caller) = $(fu{uri.user});
$var(callee) = $(ru{uri.user});
xlog("Call from $var(caller) to $var(callee)");
````

## 3. String Transformations

Length, case conversion, etc.

````kamailio-cfg
$var(name) = "Alice";

xlog("Length: $(var(name){s.len})");
xlog("Upper: $(var(name){s.upper})");
xlog("Lower: $(var(name){s.lower})");
````

## 4. String to Integer

Converting port strings to integers is a common pattern.

````kamailio-cfg
$var(port_str) = "5060";
$var(port) = $(var(port_str){s.int});

xlog("Port as string: $var(port_str)");
xlog("Port as integer: $var(port)");

if ($var(port) == 5060) {
    xlog("Standard SIP port");
}
````

## 5. Nameaddr Transformations

From/To headers use `"Display Name" <sip:user@host>;tag=xxx` format.
These transformations extract each part.

````kamailio-cfg
$var(from_hdr) = '"Alice" <sip:1001@example.com>;tag=abc123';

xlog("URI: $(var(from_hdr){nameaddr.uri})");
xlog("Name: $(var(from_hdr){nameaddr.name})");
xlog("Params: $(var(from_hdr){nameaddr.params})");
````

## 6. IP Address Transformations

Check if an IP address is valid.

````kamailio-cfg
$var(ip1) = "10.60.91.199";
$var(ip2) = "not-an-ip";

xlog("IP1 valid: $(var(ip1){ip.isip})");
xlog("IP2 valid: $(var(ip2){ip.isip})");
````

## 7. Transformation Chaining

Apply multiple transformations in sequence.

````kamailio-cfg
$var(test_uri) = "sip:alice@example.com";
xlog("Upper user: $(var(test_uri){uri.user})");
````

## 8. Base64 Encoding/Decoding

Useful for embedding binary data in SIP messages.

````kamailio-cfg
$var(original) = "Hello Kamailio";
$var(encoded) = $(var(original){s.encode.base64});
xlog("Encoded: $var(encoded)");

$var(decoded) = $(var(encoded){s.decode.base64});
xlog("Decoded: $var(decoded)");
````

---

### Transformation Summary

| Category | Transform | Description |
|----------|-----------|-------------|
| URI | `{uri.user}` | User part |
| URI | `{uri.host}` | Host part |
| URI | `{uri.port}` | Port number |
| String | `{s.len}` | Length |
| String | `{s.upper}` / `{s.lower}` | Case conversion |
| String | `{s.int}` | Convert to integer |
| Nameaddr | `{nameaddr.uri}` | Extract URI |
| Nameaddr | `{nameaddr.name}` | Extract display name |
| IP | `{ip.isip}` | IP validity check |
| Encoding | `{s.encode.base64}` | Base64 encode |
| Encoding | `{s.decode.base64}` | Base64 decode |
