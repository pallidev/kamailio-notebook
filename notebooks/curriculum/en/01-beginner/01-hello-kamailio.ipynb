## What is Kamailio?

**Kamailio** is an open-source SIP server. It handles SIP (Session Initiation Protocol) —
the core protocol for real-time communication like phone calls, video calls, and messaging.

```
📱 Phone → [SIP] → 🖥️ Kamailio → [SIP] → 📱 Phone
```

Kamailio's behavior is programmed via **.cfg files**. In this notebook we'll execute
that cfg code cell by cell so you can learn interactively.

## 1. Your First Variable Assignment

In Kamailio cfg, variables use the `$var(name)` pattern.

````python
$var(greeting) = "Hello, Kamailio!";
````

## 2. Numeric Variables

Integers work too. The type is determined automatically.

````python
%%help ds_select_dst
````

````python
$var(port) = 5060;
$var(max_calls) = 1000;
````

## 3. Printing Variables

The `xlog()` function outputs messages with variable substitution.

````python
xlog("Port: $var(port), Max calls: $var(max_calls)");
````

## 4. String Concatenation

The `+` operator concatenates values.

````python
$var(domain) = "example.com";
$var(user) = "alice";
$var(sip_uri) = "sip:" + $var(user) + "@" + $var(domain);
xlog("URI: $var(sip_uri)");
````

## 5. Inspect All Variables

The `%%vars` magic shows all currently set variables.

````python
%%vars
````

## 6. Reset State

The `%%reset` magic clears all variables.

````python
%%reset
````
