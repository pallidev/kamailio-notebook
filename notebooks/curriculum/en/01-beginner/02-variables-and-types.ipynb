## What is a SIP Message?

SIP messages are text-based, similar to HTTP:

```
INVITE sip:bob@example.com SIP/2.0
Via: SIP/2.0/UDP pc33.example.com;branch=z9hG4bK776
From: "Alice" <sip:alice@example.com>;tag=1928301774
To: <sip:bob@example.com>
Call-ID: a84b4c76e66710@pc33.example.com
CSeq: 314159 INVITE
Contact: <sip:alice@pc33.example.com>
Content-Length: 0
```

Key methods:
- **INVITE** — Start a call
- **REGISTER** — Phone registers with server
- **BYE** — End a call
- **ACK** — Acknowledge an INVITE response

## 1. Create a Mock INVITE

The `%%sip INVITE` magic creates a fake INVITE message with the same structure as a real one.

````kamailio-cfg
%%sip INVITE
From: "Alice" <sip:1001@example.com>;tag=abc123
To: <sip:1002@example.com>
Contact: <sip:1001@192.168.1.100:5060>
````

## 2. SIP Pseudo-Variables

When a SIP message is created, these variables are set automatically:

| Variable | Meaning | Example Value |
|----------|---------|---------------|
| `$ru` | Request-URI | `sip:1002@example.com` |
| `$fu` | From URI | `sip:1001@example.com` |
| `$tu` | To URI | `sip:1002@example.com` |
| `$rm` | Request Method | `INVITE` |
| `$ci` | Call-ID | (auto-generated) |
| `$ft` | From tag | `abc123` |

## 3. Read and Modify Variables

Message variables can be modified directly. In real Kamailio this is used
to route messages to different destinations.

````kamailio-cfg
# Change the Request-URI (route to a different destination)
$ru = "sip:1002@10.60.91.199:5060";
xlog("New destination: $ru");
````

## 5. Understanding Variable Types

Kamailio cfg variables have only two types:
- **string**: text values
- **integer**: whole numbers

The type is determined automatically by the value.

````kamailio-cfg
$var(str_port) = "5060";   # String "5060"
$var(int_port) = 5060;     # Integer 5060

xlog("String: $var(str_port), Integer: $var(int_port)");
````

## 6. $var vs $avp

| | `$var` | `$avp` |
|---|--------|--------|
| Scope | Process-local | Shared across transaction |
| Values | Single | Multiple (stack-like) |
| Use case | Temporary calculations, flags | Cross-transaction data |

````kamailio-cfg
$var(tmp) = "local only";
$avp(caller_id) = "1001";
$avp(caller_id) = "1002";  # Pushes onto stack

%%vars
````
